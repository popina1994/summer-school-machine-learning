```python
A = np.array([[1, 2], [3, 4]])
B = np.array([[1, 2], [3, 4]])
print(A)
print(B)
# Elementwise multiplication
print(A*B)

# Matrix multiplication
print(A @ B)
```

```python
a = np.full((3, 4), 5)
print(a)
b = np.full((3, 4), 7)
print(b)
c = np.concatenate((a, b), axis=0)
print(f"concatenate:\n {c}")
d = np.vstack((a, b))
print(f"vstack:\n {d}")
d = np.hstack((a, b))
print(f"hstack:\n {d}")

d = np.stack((a, b))
print(f"stack:\n {d}")
print(d.shape)
d = np.reshape(d, (6, 4))
print(d.shape)
print(f"reshaped:\n{d}")
# Copy
t = d.flatten()
print(f"flattened:{0}\n", t)
t[0] = 1
print(f"flattened:{0}\n", t)
# View
v = d.ravel()
print(f"ravel:{0}\n", v)
v[0] = 2
print(f"ravel:{0}\n", v)
print(d)
```
