```python
%pip install pandas
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
```

```python
import urllib.request
download_link = "https://archive.ics.uci.edu/static/public/320/student+performance.zip"
file_name = "dataset.zip"
file_name_in = "student.zip"
urllib.request.urlretrieve(download_link, file_name)
import shutil
shutil.unpack_archive(file_name, ".")
file_name_in = "student.zip"
shutil.unpack_archive(file_name_in, "./student")
```

```python
data = pd.read_csv('./student/student-mat.csv', sep=';')
data = data[[ 'age' , 'sex' , 'studytime' ,
'absences' , 'G1' , 'G2' , 'G3' ]]
print(data)
```

```python
data['sex'] = data['sex'].map({'F': 0, 'M': 1})
print(data)
```

```python
prediction = 'G3'
print(data['sex'])
print(type(data['sex']))
```

```python
#print(data)
#print(prediction)
X = np.array(data.drop([prediction], axis=1))
Y = np.array(data[prediction])
#print(f"X IS {0}", X)
#print(Y)

X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.1)
```

```python
model = LinearRegression()
model.fit(X_train, Y_train)
accuracy_train = model.score(X_train, Y_train)
accuracy = model.score(X_test, Y_test)
print(accuracy_train)
print(accuracy)
X_new = np.array([[18, 1, 3, 40, 15, 16]])
Y_new = model.predict(X_new)
print(Y_new)
```
