```python
# # Implement Fibonaci number recursively
# Fib(0) = 1
# Fib(1) = 1
# Fib(n) = Fib(n-1) + Fib(n-2)

def fib(n: int):
  if n == 0:
    return 1
  if n == 1:
    return 1
  return fib(n-1) + fib(n-2)


def fib2(n: int):
  Fib = {0:1, 1:1}
  # Fib[2] = 2
  # Fib[3] = 3

  for i in range(2, n+1):
    # print(Fib)
    Fib[i] = Fib[i-1] + Fib[i-2]
  return Fib[n]

def fib3(n: int):
  if n == 0:
    return 1
  if n == 1:
    return 1
  a = 1
  b = 1
  for i in range(2, n+1):
    c = a + b
    print("a:{} b:{} c:{}".format(a, b, c))
    a = b
    b = c
  return c
print(fib(5))
print(fib2(5))
print(fib3(5))
# Implement Fibonaci number using lists and dynamic programming
```
