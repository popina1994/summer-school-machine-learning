```python
import numpy as np
import matplotlib.pyplot as plt

x_values = np.linspace(0, 20, 100)
y_values = np.sin(x_values)
# sin x ~ x, when x ~ 0
print(x_values)
print(y_values)
```

```python
plt.plot(x_values, y_values)
plt.show()
```

```python
x = np.linspace(0, 100, 200)
y1 = 2 * x
y2 = x ** 2
y3 = np.log(x)
plt.plot(x, y1)
plt.plot(x, y2)
plt.plot(x, y3)
plt.show()
```

```python
x = np.linspace(0, 5, 200)
y1 = 2 * x
y2 = x ** 2
y3 = np.log(x)
plt.plot(x, y1,  color='r', linestyle='dashed', label='2x')
plt.plot(x, y2, color='b', linestyle='dotted', label='x^2')
plt.plot(x, y3, color='green', linestyle='dashdot', label='log x')
plt.legend(loc='upper left')
plt.show()
```

```python
x = np.linspace(0, 5, 200)
y1 = np.sin(x)
y2 = np.sqrt(x)
y3 = np.cos(x)

plt.subplot(221)
plt.plot(x, y1, 'r-')
plt.subplot(222)
plt.plot(x, y2, 'g--')
plt.subplot(223)
plt.plot(x, y3, 'b+')
plt.show()
```

```python
plt.figure(1)
plt.plot(x, y1, 'r-')
plt.figure(2)
plt.plot(x, y2, 'g-')
plt.savefig('test.png')
```

```python
mu, sigma = 200, 10
x = mu + sigma * np.random.randn(1000900)
```

```python
plt.hist(x, 1000, density=True, facecolor="blue")
plt.xlabel("height")
plt.ylabel("probability")
plt.title("Height of students")
plt.text(160, 0.125, "mu=172, sigma=4")
plt.axis([155, 220, 0, 0.15])
plt.grid(True)
```

```python
bob = ( 90 , 99 , 87 , 76 )
charles = ( 80 , 80 , 47 , 66 )
daniel = ( 40 , 95 , 76 , 89 )
skills = ( "Python" , "Java" , "Networking" , "Machine Learning")

width = 0.3
index = np.arange( 4 )
plt.bar(index, bob, width =width, label = "Bob" )
plt.bar(index + width, charles, width =width, label = "Charles" )
plt.bar(index + width * 2 , daniel, width =width, label = "Daniel" )

plt.xticks(index + width, skills)
plt.ylim( 0 , 120 )
plt.title( "IT Skill Levels" )
plt.ylabel( "Skill Level" )
plt.xlabel( "IT Skill" )
plt.legend()
```

```python
# Pie chart
labels = ( 'Pizza' , 'Hamburger' , 'French Frie' , 'Spagheti' )
values = ( 47 , 23 , 20 , 10 )
plt.pie(values, labels =labels,
autopct = "%.1f%%" , shadow = True )
plt.title( "Favorite Food" )
plt.show()
```
