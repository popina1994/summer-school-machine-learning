```python
# define your series to combine
series1 = pd.Series([4, 7, 4])
series2 = pd.Series(['data', 'science', 'dojo'])
series3 = pd.Series([1.0, 4.0, 7.0])

# combine series using pd.concat()
df = pd.concat([series1, series2, series3], axis = 1)
print(df)
```

```python
df_first = pd.DataFrame({'key': [0, 1, 2, 3, 4, 5],
                    'A': ['A0', 'A1', 'A2', 'A3', 'A4', 'A5']})
df_first.set_index('key')

other = pd.DataFrame({'key': [0, 1, 2, 6],
                       'B': ['B0', 'B1', 'B2', 'B6']})
other.set_index('key')

print(df_first)
print(other)
#df_first.join(other, lsuffix='first',how='inner', on='key', rsuffix='_other')
print("Inner join")
print(pd.merge(df_first, other, how='inner', on='key'))
print("Left join", pd.merge(df_first, other, how='left', on='key'))
print("Right join", pd.merge(df_first, other, how='right', on='key'))
```
